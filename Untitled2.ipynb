```python
import cv2
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # or your trained model
video_path = "sample_video.mp4"

# Open input video to get properties
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
cap.release()

# Define output writer
out = cv2.VideoWriter("output.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

# Run detection with streaming
results = model.predict(source=video_path, stream=True)

for r in results:
    frame = r.plot()  # Draw boxes and labels on the frame
    out.write(frame)  # Write frame to output video

out.release()
print("✅ Video saved as output.mp4")


```
